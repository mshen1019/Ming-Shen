# Analysis of the relationship between Covid-19 daily new cases and political leaning of US states 

```python
import pandas as pd
import numpy as np
```

load COVID-19 data downloaded from CDC
column "new_case" stands for daily new cases
column "submission_date" stands for the date of report date of that data point
The "submission_date" ranges from Jan 29th, 2020 to Feb 1st, 2021

```python
covid=pd.read_csv("United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv")
```

```python
covid.head()
```

load US population data
column "party" stands for the result of US President Election 2020 of that state

```python
population=pd.read_csv("popultion.txt",sep="\t")
```

```python
population.head()
```

Drop The column "Unnamed: 0"

```python
population_1=population.drop("Unnamed: 0",axis=1)
```

```python
population_1.head()
```

Merge dataset "population_1" and "covid" by column "state"

```python
covid_pop=pd.merge(population_1,covid, on=["state"])
```

Keep only five columns: "state","population","party","submission_date","new_case"

```python
covid_pop_1=covid_pop[["state","population","party","submission_date","new_case"]]
```

Generate new dataset covid_pop_2 with column "new_case_per_million" and drop the columns "new_case","population" 

```python
covid_pop_1["new_case_per_million"]=covid_pop_1["new_case"]/covid_pop_1["population"]*1000000
```

```python
covid_pop_2=covid_pop_1.drop(["new_case","population"],axis=1)
covid_pop_2.head()
```

Unstack column 'submission_date'

```python
covid_pop_3=covid_pop_2.set_index(['state','party', 'submission_date']).unstack('submission_date')
covid_pop_3.head()
```

Calculate the 7-day average daily new cases per million people and use it as input data

convert dataset into matrix

```python
covid_pop_array=covid_pop_3.values
```

```python
len(covid_pop_array)
```

```python
len(covid_pop_array[0])
```

Use average_matrix to generate (2n+1)-day average values of each element in the matrix.
We use n=3 here.

```python
def average_matrix(mat,n):
    x=len(mat)
    y=len(mat[0])
    new_mat=np.zeros((x,y))
    for i in range(x):
        for j in range(n,y-n):
            sum_all=mat[i,j]
            for k in range(1,n+1):
                sum_all=sum_all+mat[i,j-n]+mat[i,j+n]
            new_mat[i,j]=sum_all/7
    return new_mat
```

```python
new_covid_pop_array=average_matrix(covid_pop_array,3)
```

Use fill_df to fill the values in the matrix into original dataset

```python
def fill_df(df,mat):
    df_new=df
    x=len(mat)
    y=len(mat[0])
    for i in range(x):
        for j in range(y):
            df_new.iloc[i,j]=mat[i,j]
    return df_new
```

```python
new_covid_pop_df=fill_df(covid_pop_3,new_covid_pop_array)
```

We would not use the data from first three days and last three days, so just let these be 0
